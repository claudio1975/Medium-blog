```python
from datasets import load_dataset
import pandas as pd
from transformers import pipeline
import re
```

```python

# Load dataset
dataset = load_dataset("OpenMed/Medical-Reasoning-SFT-GPT-OSS-120B", split="train")

```

```python


# Convert to DataFrame for easier searching
data = []
for example in dataset:
    messages = example['messages']
    if messages:
        question = messages[0]['content'] if messages[0]['role'] == 'user' else ""
        answer = messages[1]['content'] if len(messages) > 1 and messages[1]['role'] == 'assistant' else ""
        data.append({'question': question, 'answer': answer})

df = pd.DataFrame(data)

```

```python
# Now you can easily search
print("\nSearching for disease symptoms...\n")

# Example: Find all questions about heart attack
heart_attack_df = df[df['question'].str.contains('heart attack|myocardial infarction|chest pain', case=False, na=False)]
#print(f"Found {len(heart_attack_df)} questions about heart attacks")
heart_attack_df
```

```python
pipe = pipeline("text-generation", model="towardsinnovationlab/qwen3-medical")
```

```python
message1 = [
    {"role": "user", "content": "In the context of the human cardiovascular system, consider a scenario where a patient has a genetic mutation that leads \
    to a significant reduction in the production of nitric oxide (NO) by endothelial cells. This mutation affects the dilation of blood vessels and the \
    regulation of vascular tone. Which of the following cardiovascular conditions is most likely to be a direct consequence of this mutation? \
    A Atrial fibrillation \
    B Congestive heart failure \
    C Hypertension \
    D Myocardial infarction \
    E Peripheral artery disease \
    F Pulmonary edema \
    G Thromboembolic disease \
    H Atherosclerosis \
    I Endocarditis \
    J Vasculitis"},
]
result1 = pipe(message1)
# Extract only the assistant's response
assistant_response1 = result1[0]['generated_text'][-1]['content']
# Remove the <think> tags and their content
clean_response1 = re.sub(r'<think>.*?</think>', '', assistant_response1, flags=re.DOTALL).strip()
print(clean_response1)
```

```python
message2 = [
    {"role": "user", "content": "Which of the following hormones is primarily responsible for the regulation of blood glucose levels and is secreted by the pancreas? \
    A: Insulin B: Glucagon C: Parathyroid hormone (PTH) D: Calcitonin \
    E: Thyroxine (T4) F: Cortisol G: Estrogen H: Progesterone I: Oxytocin J: Vasopressin (ADH)"},
]
result2 = pipe(message2)
# Extract only the assistant's response
assistant_response2 = result2[0]['generated_text'][-1]['content']
# Remove the <think> tags and their content
clean_response2 = re.sub(r'<think>.*?</think>', '', assistant_response2, flags=re.DOTALL).strip()
print(clean_response2)
```
