```python
def evaluate_answers(question, reference, answer_a, answer_b, answer_c):
    prompt=f"""
    You are a a neutral evaluator. Your task is to asses three candidate answers to a question by comparing each to a reference answer.
    Evaluate based on factual accuracy, completeness, clarity, and alignement with the reference.

    ---
    **Question:** {question}
    **Reference answer:** {reference}
    **Answer A:** {answer_a}
    **Answer B:** {answer_b}
    **Answer C:** {answer_c}
    ---
    Please provide:
    1. A brief evaluation of each candidate answer.
    2. A final judjement: which answer (A,B,or C) is most aligned with the reference and why?
    """
    response = client.chat.completions.create(
        model="gpt-5.2",
        messages=[
            {"role": "user","content": prompt}],
            temperature=0.2,
    )

    return response.choices[0].message.content
```
